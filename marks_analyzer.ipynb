```python
def calculate_average(marks):
    return sum(marks) / len(marks)

marks = []

for i in range(3):
    m = int(input("Enter marks: "))
    marks.append(m)

avg = calculate_average(marks)

print("Average marks:", avg)

if avg >= 50:
    print("Status: Pass")
else:
    print("Status: Fail")
```
