```python
def study_message(name, hours):
    if hours >= 2:
        return "Good job " + name
    else:
        return "Try to study more " + name

print(study_message("Sadaf", 3))
```

```python
def show_tasks(task_list):
    for task in task_list:
        print("-", task)

tasks = []
tasks.append("Python practice")
tasks.append("GitHub commit")
tasks.append("Mini project")

show_tasks(tasks)
```
